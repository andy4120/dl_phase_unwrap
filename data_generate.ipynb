```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = './000.bmp'
img = cv2.imread(img_path)
print(img.shape)

plt.imshow(img)
plt.show()
```

```python
# img flip
img = cv2.flip(img, 1)

plt.imshow(img)
plt.show()
```

```python
x = 930
y = 600
img_crop = img[y:y+512, x:x+512]

plt.imshow(img_crop)
plt.show()
```

```python
# reduce [0, 255] -> [0, 127]
eye_norm = cv2.normalize(img_crop, None, 0, 127, cv2.NORM_MINMAX)

plt.imshow(eye_norm)
plt.show()
```

```python
# reduce single shot [0, 255] -> [0, 127]
sim_path = './dl_data_set/dl_deflec_eye/00006/img_8.png'
single_shot = cv2.imread(sim_path)
print(single_shot.shape)

sim_norm = cv2.normalize(single_shot, None, 0, 127, cv2.NORM_MINMAX)

plt.imshow(sim_norm)
plt.show()
```

```python
# overlay images
overlay = eye_norm + sim_norm

plt.imshow(overlay)
plt.show()
```
